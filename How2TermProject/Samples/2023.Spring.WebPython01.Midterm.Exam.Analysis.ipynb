## 4.5. (H5) Student who finished all 12 Slido Quizzes got higher Quiz score.

## Slido Quiz and Mid-term Exam Score

```python
plt.figure(figsize=(10,8))
plt.xticks(range(0, 13))
plt.xlim([-1, 13])

g = sns.histplot(
    data=data, x="Slido 중간고사 복습", 
    bins=np.arange(-1,13) + 0.5, 
)
g.set_xlabel("The Number Of Slido Quiz that Studied")
g.set_ylabel("The number of Students")
```

```python
plt.figure(figsize=(10,8))
plt.yticks(np.arange(0, 21, 5))

def slido_categorizer(num):
    if num in (12,):
        return '12'
    else:
        return '0~11'

data['Slido_Group']=data['Slido 중간고사 복습'].apply(slido_categorizer)

g =sns.boxplot(
    data=data,
    x="Slido_Group", 
    y="중간고사 퀴즈",
    #order=['0~8', '9~10', '10~12'],
)
g.set_xlabel("The Number Of Slido Quiz that Studied")
g.set_ylabel("The number of Students")

plt.yticks(range(0, 11))
plt.show()
```

```python
slido_df = pd.DataFrame.copy(data[['학번']])

for i in range(1, 13):
    df = pd.read_csv(f'.\\slido\\{str(i).zfill(2)}. Slido Quiz.csv')   
    df = df[df["시도"]==1][["SIS ID", "제출됨"]] # 첫 번째 시도의 제출됨 날짜값만 남기고 나머지 행 다 지우기
    df.rename(columns = {'SIS ID' : '학번', '제출됨': f'Slido{i}'}, inplace = True)
    
    df[f'Slido{i}'] = pd.to_datetime(df[f'Slido{i}']).dt.tz_convert(tz='Asia/Seoul').dt.tz_localize(None)
    df[f'Slido{i}'] = pd.to_datetime(df[f'Slido{i}'], format='%Y-%m-%d %H:%M:%S')
    slido_df = pd.merge(slido_df, df, on="학번", how='left')

slido_df['all_clear'] = ~slido_df.iloc[:, 1:].apply(np.isnat, axis=1).any(axis=1)

#slido_df
```

```python
slido_analyzer = pd.DataFrame(
    {
        'date': pd.date_range("2023-04-18", "2023-04-25")
    }
)

def get_good_student_counter(date):
    """The number of students who studied all Slido Quizzes until each date
    """
    students = slido_df.iloc[:,1:-1].apply(lambda submit: submit.dt.date<=date, axis=1).apply(all, axis=1)
    return sum(students)
    
slido_analyzer['cumulative_sum'] = slido_analyzer['date'].apply(get_good_student_counter)

slido_analyzer
```

```python
plt.figure(figsize=(10,8))
g = sns.lineplot(
    data=slido_analyzer, 
    x="date", 
    y="cumulative_sum", 
    marker= 'o', 
    markersize=14)

g.set_xlabel('')
g.set_ylabel("The Cumulative Sum of Students")

plt.yticks(range(0, 131, 10))
plt.xticks(ticks=slido_analyzer['date'], labels=[f'04-{i}' for i in range(18, 26)])
plt.show()
```

```python
slido_all_clear = slido_df[slido_df['all_clear']]
slido_all_clear['last_clear_date'] = slido_all_clear.iloc[:, 1:-1].max(skipna=True, axis=1)
slido_all_clear['last_clear_day'] = slido_all_clear['last_clear_date'].dt.day
slido_all_clear['last_clear_hour'] = slido_all_clear['last_clear_date'].dt.hour
slido_all_clear.to_excel("./slido_access_info.xlsx")


lastday_graph = slido_all_clear[slido_all_clear['last_clear_day']==25].groupby('last_clear_hour')['학번'].count().reset_index()
lastday_graph
```

```python
plt.figure(figsize=(10,8))
plt.xticks(range(0, 13))
plt.yticks(range(0, 11))
plt.xlim([-1, 13])
plt.ylim([0, 11])

g = sns.barplot(data=lastday_graph, x="last_clear_hour", y="학번", color='b')
g.set_xlabel("25th April - Hour")
g.set_ylabel("# of Students Who Finished Studying All Slido Quizzes")
plt.show()
```
