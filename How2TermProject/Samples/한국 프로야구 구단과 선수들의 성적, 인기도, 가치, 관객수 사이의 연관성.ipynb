```python
import time
import math
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd    
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

KBO_team_name = ["삼성", "KIA", "롯데", "LG", "두산", "한화", "SSG", "키움", "NC", "KT"]

# matplotlib 한글 폰트 설정
fm.get_fontconfig_fonts()
font_location = 'Fonts/malgun.ttf' # For Wi
font_name = fm.FontProperties(fname=font_location).get_name()
plt.style.use('default')
plt.rcParams['font.family'] = font_name
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 10

# 1. 구단별 구단가치
tbl_team_value_df = pd.read_csv("./구단가치.csv", encoding='utf-8')

# 2. 구단별 리그순위
options = webdriver.ChromeOptions()
options.add_argument("headless")
driver = webdriver.Chrome('chromedriver_win32/chromedriver', options=options)

driver.get("https://www.koreabaseball.com/TeamRank/TeamRank.aspx")

tbl_udpRecord_datalist = [[0 for j in range(0, 6)] for i in range(0, 10)]
tbl_udpRecord_datalist_average = {'삼성': 0, 'KIA': 0, '롯데': 0, 'LG': 0, '두산': 0,
                                  '한화': 0, 'SSG': 0, '키움': 0, 'NC': 0, 'KT': 0}
for index in range(2, 9):
    if index != 3:
        driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_ddlYear"]/option[' + str(index) + ']').click()
        time.sleep(0.1)
        for rank in range(1, 11):
            data = driver.find_element_by_xpath(
                '//*[@id="cphContents_cphContents_cphContents_udpRecord"]/table/tbody/tr[' + str(rank) + ']/td[2]').text
            if data == 'SK':
                data = 'SSG'
            elif data == '넥센':
                data = '키움'
            tbl_udpRecord_datalist_average[data] += rank
            tbl_udpRecord_datalist[KBO_team_name.index(data)][(index - 2) if (index == 2) else (index - 3)] = rank

driver.quit()

for key in tbl_udpRecord_datalist_average.keys():
    tbl_udpRecord_datalist_average[key] = round(tbl_udpRecord_datalist_average[key] / 6, 1)

tbl_udpRecord_df = pd.DataFrame(data=tbl_udpRecord_datalist,
                                columns=['2021년', '2019년', '2018년', '2017년', '2016년', '2015년'])
tbl_udpRecord_df.insert(0, '구단명', KBO_team_name)
tbl_udpRecord_df['평균'] = tbl_udpRecord_datalist_average.values()
```

```python
# 1. 구단별 리그순위 - 가설1에서 분석한 결과를 재사용한다.

# 2. 구단별 인기도
req = requests.get('https://www.koreabaseball.com/History/Crowd/History.aspx')
html = req.text
soup = BeautifulSoup(html, 'html.parser')

tbl_crowd_history_datalist = []
for index in range(1, 8):
    if index != 2:
        tbl_crowd_history_datalist.append(
            soup.select('#tblHistory > tbody > tr:nth-of-type(' + str(index) + ')')[0].text.replace(' ', '').split()[1:11])

tbl_crowd_history_average = []
tbl_crowd_history_datalist_fixed = [[0 for j in range(0, 6)] for i in range(0, 10)]

for i in range(0, 10):
    result = 0
    data_temp = ''
    for j in range(0, 6):
        data = tbl_crowd_history_datalist[j][i][tbl_crowd_history_datalist[j][i].find('(') + 1:-1].replace(',', '')
        tbl_crowd_history_datalist_fixed[i][j] = data
        result += int(data)
    tbl_crowd_history_average.append(math.trunc(result / 6))

tbl_crowd_history_df = pd.DataFrame(data=tbl_crowd_history_datalist_fixed,
                                    columns=['2021년', '2019년', '2018년', '2017년', '2016년', '2015년'])
tbl_crowd_history_df.insert(0, '구단명', KBO_team_name)
tbl_crowd_history_df['평균'] = tbl_crowd_history_average
```

> 1. 코로나로 인한 무관중기간(2020년,2021년)과 관중이 있었을 때의 기간(2018년,2019년)별 정규리그에서 발생한 프로야구 선수의 홈런, 안타의 평균을 구한다. 이때 출전선수의 수가 매해 다르므로 최소값인 53명으로 고정하여 평균을 구한다. <br>

> * **데이터 획득 및 분석 도구**

```python
# 관중기간 VS 무관중기간
options = webdriver.ChromeOptions()
options.add_argument("headless")
driver = webdriver.Chrome('chromedriver_win32/chromedriver', options=options)

driver.get("https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx?sort=HRA_RT")

tbl_player_record_columns = ['안타', '홈런']
tbl_player_record_datalist = [[0, 0], [0, 0]]

for index in range(37, 41):
    driver.find_element_by_xpath(
        '//*[@id="cphContents_cphContents_cphContents_ddlSeason_ddlSeason"]/option[' + str(index) + ']').click()  # 2018년
    driver.find_element_by_xpath(
        '//*[@id="cphContents_cphContents_cphContents_ddlSeries_ddlSeries"]/option[1]').click()  # KBO 정규시즌
    driver.find_element_by_xpath(
        '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo1"]').click()  # 1페이지
    time.sleep(1.5)

    for number in range(1, 31):
        H = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[9]').text
        HR = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[12]').text
        AB = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[7]').text

        tbl_player_record_datalist[0 if index < 39 else 1][0] += round(int(H) / int(AB), 1)
        tbl_player_record_datalist[0 if index < 39 else 1][1] += round(int(HR) / int(AB), 1)

    driver.find_element_by_xpath(
        '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo2"]').click()  # 2페이지
    time.sleep(1.5)

    for number in range(1, 24):
        H = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[9]').text
        HR = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[12]').text
        AB = driver.find_element_by_xpath(
            '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[3]/table/tbody/tr[' + str(number) + ']/td[7]').text

        tbl_player_record_datalist[0 if index < 39 else 1][0] += round(int(H) / int(AB), 1)
        tbl_player_record_datalist[0 if index < 39 else 1][1] += round(int(HR) / int(AB), 1)

driver.quit()

tbl_player_record_df = pd.DataFrame(data=tbl_player_record_datalist,
                                    columns=tbl_player_record_columns)
tbl_player_record_df.insert(0, '구분', ['관중기간', '무관중기간'])
```
