## 1. 주제 선정 이유
- 교통사고는 한순간으로 사람의 목숨을 앗아갈 수 있는 아찔한 사고이다. 우리는 날마다 횡단보도를 건너고, 대중교통을 이용하기도 하며 사고의 위험에 노출된다. 기상, 노면 상태별 교통사고 현황을 분석해 봄으로써 특별히 조심해야 되는 날씨나 도로가 있는지 알아본다.

## 2. 가설 정의

- 가설1. 기상상태별(맑음, 흐림, 비, 안개, 눈, 기타) 교통사고 사망률이<br>'맑음 < 흐림 < 안개 < 비 < 눈' 순으로 높을 것이다.</br><br>

- 가설2. 기상상태별 부상률과 사망률은 비례관계일 것이다. </br>
<br>

- 가설3. 노면상태별(건조, 젖음/습기, 서리/결빙, 적설, 기타) 교통사고 사망률이<br> '건조 < 젖음/습기 < 적설 < 서리/결빙' 순으로 높을 것이다. <br>(사망률은 (사망자수)/(사고건수)로 계산한다)</br>
<br>

- 가설4. 노면상태별 부상률과 사망률은 비례관계일 것이다.</br>
<br>

- 가설5. 연중 교통사고 사고 건수는 월별로 봤을 때 겨울(12~2월)에 가장 많을 것이다.</br>
<br>

- 가설6. 연중 교통사고 사망률은 겨울(12~2월)에 가장 높을 것이다.</br>
<br>

- 가설7. 연중 교통사고 부상률은 겨울(12~2월)에 가장 높을 것이다.

</br>※사망률은 (사망자수)/(사고건수)로 계산한다<br>
※부상률은 (부상자수)/(사고건수)로 계산한다<br>

### 4-1. 기상상태별 교통사고 치명률을 계산하기 위한 데이터 가공

```python
import pandas as pd
# 날씨 데이터 분석을 위한 dataframe 선언 및 초기화
df_weather = pd.read_csv('./data/도로교통공단_월별 기상상태별 교통사고 현황_20191231.csv', encoding = 'euc-kr')

# 날씨 데이터확인
df_weather
```

#### 기본 dataframe 만들기

```python
# NaN으로 채워진 행과 열 삭제
df_weather = df_weather.drop([0,1,2,3,16])
df_weather = df_weather.loc[:, :df_weather.columns[21]]

# 누락된 데이터(NaN)를 '0'으로 대체
df_weather = df_weather.fillna('0')

# 데이터를 정수형으로 변환
for i in range(22):
    df_weather[f'Unnamed: {i}'] = df_weather[f'Unnamed: {i}'].str.replace(",", "").astype(int)
    
# 날씨별 dataframe 만들기
df_sunny = df_weather.loc[4: , : df_weather.columns[3]]
df_cloudy = df_weather.loc[4: , df_weather.columns[4] : df_weather.columns[6]]
df_rainy = df_weather.loc[4: , df_weather.columns[7] : df_weather.columns[9]]
df_foggy = df_weather.loc[4: , df_weather.columns[10] : df_weather.columns[12]]
df_snowy = df_weather.loc[4: , df_weather.columns[13] : df_weather.columns[15]]
df_etc = df_weather.loc[4: , df_weather.columns[16] : df_weather.columns[18]]
df_total = df_weather.loc[4: , df_weather.columns[19] : df_weather.columns[21]]

# 날씨별 dataframe에 발생월 데이터 추가하기
df_month = df_weather.loc[4: , df_weather.columns[0]]
df_cloudy = pd.concat([df_month, df_cloudy], axis = 1)
df_rainy = pd.concat([df_month, df_rainy], axis = 1)
df_foggy = pd.concat([df_month, df_foggy], axis = 1)
df_snowy = pd.concat([df_month, df_snowy], axis = 1)
df_etc = pd.concat([df_month, df_etc], axis = 1)
df_total = pd.concat([df_month, df_total], axis = 1)

# rows 인덱스 초기화
df_sunny = df_sunny.reset_index()
del df_sunny['index']
df_cloudy = df_cloudy.reset_index()
del df_cloudy['index']
df_rainy = df_rainy.reset_index()
del df_rainy['index']
df_foggy = df_foggy.reset_index()
del df_foggy['index']
df_snowy = df_snowy.reset_index()
del df_snowy['index']
df_etc = df_etc.reset_index()
del df_etc['index']
df_total = df_total.reset_index()
del df_total['index']

# columns 이름 설정하기
df_sunny.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_cloudy.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_rainy.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_foggy.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_snowy.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_etc.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_total.columns = ['발생월', '사고건수', '사망자수', '부상자수']

# 사망률 column 추가하기
df_sunny['사망률'] = (df_sunny['사망자수'] / df_sunny['사고건수']) * 100
df_cloudy['사망률'] = (df_cloudy['사망자수'] / df_cloudy['사고건수']) * 100
df_rainy['사망률'] = (df_rainy['사망자수'] / df_rainy['사고건수']) * 100
df_foggy['사망률'] = (df_foggy['사망자수'] / df_foggy['사고건수']) * 100
df_etc['사망률'] = (df_etc['사망자수'] / df_etc['사고건수']) * 100
df_total['사망률'] = (df_total['사망자수'] / df_total['사고건수']) * 100
    # snowy는 사고건수가 0인 경우가 있어서 zero division error가 나기 때문에
    # 사고건수에 1을 대체한 후 사망률 column을 추가하고 다시 사고건수를 원래대로 복원함
df_snowy['사고건수'].replace(0, 1, inplace=True)
df_snowy['사망률'] = (df_snowy['사망자수'] / df_snowy['사고건수']) * 100
df_snowy['사고건수'].replace(1, 0, inplace=True)

# 부상률 column 추가하기
df_sunny['부상률'] = (df_sunny['부상자수'] / df_sunny['사고건수']) * 100
df_cloudy['부상률'] = (df_cloudy['부상자수'] / df_cloudy['사고건수']) * 100
df_rainy['부상률'] = (df_rainy['부상자수'] / df_rainy['사고건수']) * 100
df_foggy['부상률'] = (df_foggy['부상자수'] / df_foggy['사고건수']) * 100
df_etc['부상률'] = (df_etc['부상자수'] / df_etc['사고건수']) * 100
df_total['부상률'] = (df_total['부상자수'] / df_total['사고건수']) * 100
    # snowy는 사고건수가 0인 경우가 있어서 zero division error가 나기 때문에
    # 사고건수에 1을 대체한 후 사망률 column을 추가하고 다시 사고건수를 원래대로 복원함
df_snowy['사고건수'].replace(0, 1, inplace=True)
df_snowy['부상률'] = (df_snowy['부상자수'] / df_snowy['사고건수']) * 100
df_snowy['사고건수'].replace(1, 0, inplace=True)
```

### 4-2. 노면상태별 교통사고 치명률을 계산하기 위한 데이터 가공

```python
# 노면 상태 데이터 분석을 위한 dataframe 선언 및 초기화
df_surface = pd.read_csv('./data/도로교통공단_월별 노면상태별 교통사고 현황_20191231.csv', encoding = 'euc-kr')

# 노면 데이터 확인
df_surface
```

#### 데이터 정제하기

```python
# NaN으로 채워진 행 삭제
df_surface = df_surface.drop([0,1,2,3])

# 데이터가 너무 적어 유의미한 통계를 낼 수 없는 비포장 상태의 데이터 제거
df_surface = df_surface.loc[:, :df_surface.columns[21]]

# 누락된 데이터(NaN)를 '0'으로 대체
df_surface = df_surface.fillna('0')

# columns 이름 통일
df_surface.rename(columns={'*노면상태별 월별 교통사고(2019년)':'Unnamed: 0'}, inplace=True)

# 데이터를 정수형으로 변환
for i in range(22):
    df_surface[f'Unnamed: {i}'] = df_surface[f'Unnamed: {i}'].str.replace(",", "").astype(int)

# 노면상태별 dataframe 만들기
# 침수와 해빙은 유의미한 통계를 낼 만큼의 데이터가 없으므로 제외
df_dry = df_surface.loc[4: , : df_surface.columns[3]]
df_wet = df_surface.loc[4: , df_surface.columns[4] : df_surface.columns[6]]
df_ice = df_surface.loc[4: , df_surface.columns[7] : df_surface.columns[9]]
df_snow = df_surface.loc[4: , df_surface.columns[10] : df_surface.columns[12]]
df_surface_etc = df_surface.loc[4: , df_surface.columns[13] : df_surface.columns[15]]

# 노면상태별 dataframe에 발생월 데이터 추가하기
df_month = df_surface.loc[4: , df_surface.columns[0]]
df_wet = pd.concat([df_month, df_wet], axis = 1)
df_ice = pd.concat([df_month, df_ice], axis = 1)
df_snow = pd.concat([df_month, df_snow], axis = 1)
df_surface_etc = pd.concat([df_month, df_surface_etc], axis = 1)

# rows 인덱스 초기화
df_dry = df_dry.reset_index()
del df_dry['index']
df_wet = df_wet.reset_index()
del df_wet['index']
df_ice = df_ice.reset_index()
del df_ice['index']
df_snow = df_snow.reset_index()
del df_snow['index']
df_surface_etc = df_surface_etc.reset_index()
del df_surface_etc['index']

# columns 이름 설정하기
df_dry.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_wet.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_ice.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_snow.columns = ['발생월', '사고건수', '사망자수', '부상자수']
df_surface_etc.columns = ['발생월', '사고건수', '사망자수', '부상자수']

# 사망률 column 추가하기
df_dry['사망률'] = (df_dry['사망자수'] / df_dry['사고건수']) * 100
df_wet['사망률'] = (df_wet['사망자수'] / df_wet['사고건수']) * 100
df_ice['사망률'] = (df_ice['사망자수'] / df_ice['사고건수']) * 100
df_snow['사망률'] = (df_snow['사망자수'] / df_snow['사고건수']) * 100
df_surface_etc['사망률'] = (df_surface_etc['사망자수'] / df_surface_etc['사고건수']) * 100
# snow랑 ice는 사고건수가 0인 경우가 있어서 zero division error가 나기 때문에
    # 사고건수에 1을 대체한 후 사망률 column을 추가하고 다시 사고건수를 원래대로 복원함
df_snow['사고건수'].replace(0, 5, inplace=True)
df_snow['사망률'] = (df_snow['사망자수'] / df_snow['사고건수']) * 100
df_snow['사고건수'].replace(5, 0, inplace=True)

df_ice['사고건수'].replace(0, 4, inplace=True)
df_ice['사망률'] = (df_ice['사망자수'] / df_ice['사고건수']) * 100
df_ice['사고건수'].replace(4, 0, inplace=True)

# 부상률 column 추가하기
df_dry['부상률'] = (df_dry['부상자수'] / df_dry['사고건수']) * 100
df_wet['부상률'] = (df_wet['부상자수'] / df_wet['사고건수']) * 100
df_ice['부상률'] = (df_ice['부상자수'] / df_ice['사고건수']) * 100
df_snow['부상률'] = (df_snow['부상자수'] / df_snow['사고건수']) * 100
df_surface_etc['부상률'] = (df_surface_etc['부상자수'] / df_surface_etc['사고건수']) * 100
# snow랑 ice는 사고건수가 0인 경우가 있어서 zero division error가 나기 때문에
    # 사고건수에 1을 대체한 후 사망률 column을 추가하고 다시 사고건수를 원래대로 복원함
df_snow['사고건수'].replace(0, 5, inplace=True)
df_snow['부상률'] = (df_snow['부상자수'] / df_snow['사고건수']) * 100
df_snow['사고건수'].replace(5, 0, inplace=True)

df_ice['사고건수'].replace(0, 4, inplace=True)
df_ice['부상률'] = (df_ice['부상자수'] / df_ice['사고건수']) * 100
df_ice['사고건수'].replace(4, 0, inplace=True)
```
