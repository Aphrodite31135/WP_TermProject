이 그래프는 전동킥보드 가해자 연령대를 분석하기 위해 x축에는 년도를, y축에는 사고 건수를 나태내고 각 년도에 연령대를 나누어 출력되게 하였다. 2017년부터 2019년까지의 사고 건수를 분석해보았다. 처음에 x축을 연령대로 지정하였더니, 매년 발생하는 사고 건수가 다르기에 상대적으로 비교하는 것이 어려웠다. 가설을 증명하기 위해서는 같은 년도에 발생한 사고 가해자 중 어떤 연령대가 가장 많은지 파악해야 하기에 결국 x축을 년도로 지정해주게 된 것이다.  

그래프를 통해 알 수 있듯이, 모든 년도에서 가해자는 20대가 가장 많다. 또한 생각보다 30대 이상도 많이 있다는 것을 알 수 있었으며, 10대의 사고 건수도 꽤 있는 것을 파악할 수 있다.  

전동킥보드를 운행하기 위해서는 운전면허증 인증이 필요하기에 사실상 10대의 사고 건수는 0으로 나오는 것이 옳다. 하지만 사고 건수가 0은커녕 점차 증가하는 양상을 띤다. 무면허 운전이기에 10대들이 운전하면 사고가 나는 것은 당연한 일이다. 또한, 뉴스나 주변 길거리만 보더라도 10대들이 전동킥보드를 운전하고 다니는 것을 쉽게 확인할 수 있다. 전동킥보드 회사들도 10대들의 운전 사실을 알고 있을 텐데, 왜 아직까지 개선되지 않는 것일까. 

찾아본 결과, 운전 면허증 인증 없이 무면허 청소년에게 이동장치를 대여하더라도 현행법상 처벌이 불가능하다는 것을 알아냈다. 아마 전동킥보드 회사들은 회사의 이익만 생각하여 10대 청소년들의 무면허 운전을 방관하고 있는 것으로 추정된다. 우리는 앞선 가설의 증명을 통해 앞으로 전동킥보드의 사용률 및 사고 건수가 매우 증가할 것임을 알아냈다. 이는 즉, 10대 청소년들의 사용률 또한 증가할 것이라고 추측할 수 있다. 이 그래프만 봐도 10대들의 사고 건수는 증가하고 있기 때문이다.  

따라서 사고 건수를 줄이기 위해서는 회사들의 빠른 대처도 필요하다. 그들은 자신들의 이익을 우선시하는게 아닌 청소년들의 안전을 생각하여, 10대들의 무면허 운전을 방치하지 말아야한다. 즉, 각 회사들은 운전면허 인증 절차가 부실한 것을 깨닫고, 이를 개선시키기 위한 노력을 해야 한다. 또한, 무면허 개인형 이동장치 이용을 방관하는 공유업체를 규제할 법 개정도 매우 필요해보인다. 또한 10대들은 오류로 인해 운전면허 인증 절차를 통과하였더라도, 탑승하지 말아야 한다.

"가해자는 20대가 가장 많고, 10대 또한 사고를 많이 낸다."라는 가설이 사실이란 것을 증명해보았다. 전동킥보드 사용자들의 문제뿐만 아니라 회사에도 문제가 있다는 것을 깨닫게 되는 분석이었다. 2021년까지의 데이터를 분석했다면, 증가하는 양상이 더 크게 보여 한눈에 파악하기 쉬웠을 텐데, 데이터를 찾지 못하여 2019년까지만 출력하였다는 점이 아쉬웠다.  

### 2-1) 가설 분석
- 불법 주,정차된 전동킥보드가 많을 것이다.

```python
#import pandas as pd
#import matplotlib.pylab as plt

df=pd.read_excel('./서울시 전동킥보드 견인 현황.xlsx')

df1=df.loc[0:60604,['신고일시']]
df1['year']=df['신고일시'].dt.year     #신고일시에서 year을 뽑아서 데이터 만들기
df1
```

```python
def count_2021():      #2021년도에 견인된 전동킥보드 개수 구하는 함수
    x=0
    for i in range(0,60605):    #60605번째를 포함안해
        if df1['year'].loc[i]==2021:
            x+=1
        else:
            continue
    return x

def count_2022():      #2021년도에 견인된 전동킥보드 개수 구하는 함수
    y=0
    for i in range(0,60605):    
        if df1['year'].loc[i]!=2021:
            y+=1
        else:
            continue
    return y

a=count_2021()
b=count_2022()
c=[a,b]
year=['2021','2022']

fig, ax = plt.subplots(figsize=(7,5))

plt.ylim(0,40000)
plt.xlabel('<년도>',size=12)
plt.ylabel('<견인건수>',size=12)


plt.bar(['2021','2022'],[a,b],color='indianred',width=0.6)
plt.title('<년도별 견인된 전동킥보드 수>',size=15)

#plt.grid(True,axis='y')             #y축에만 그리드 설정
plt.show()
```

2021년 07월부터 2022년 07월까지의 데이터를 가지고 분석해보았다. 엑셀 파일의 신고일시 데이터 유형이 object였기에 년도 데이터만 빼서 원래 표에 'year'이라는 새로운 데이터로 추가하였다.  

이후 2021년, 2022년 견인 건수를 파악하기 위해 엑셀 표를 전부 읽으면서 year이 2021년일 때 1을 추가하여 2021년의 개수를 세는 함수 count_2021과 똑같은 방식으로 2022의 개수를 세는 함수 count_2022를 만들었다. 이를 통해 나온 결과값으로 표를 만들었고, 이를 통해 견인 건수가 2021년에는 6개월에 20,000건, 2022년에는 7개월에 40,000건이라는 것을 알 수 있었다. 즉, 2022년의 견인 건수는 2021년의 약 2배 정도 더 많다. 

```python
a1=20000/(30*6)     #2021년 하루당 견인건수
a2=40000/(30*7)     #2022년 하루당 견인건수
print(a1,a2)
```

이렇게 어림잡아 계산해본 결과 2021년에는 하루당 약 111대가, 2022년에는 하루당 약 190대가 견인되었다는 것을 알 수 있다. 따라서 "불법 주,정차된 전동킥보드가 많을 것이다"라는 가설은 옳았다.  

이를 통해 주차 구역 설정에 대한 필요성이 크다는 것을 알 수 있다. 견인되기까지 평균 일수가 이틀정도였는데, 만약 하루나 이틀정도 안에 견인이 되지 않는다면, 이로 인한 추가 사고가 발생할 수도 있을 것이다. 

하지만 시간이 지나면서 사용률이 증가했기 때문에 '견인 건수가 증가하고 있다'라고 말하기는 어렵다. 데이터를 2021년에는 하반기, 2022년에는 상반기를 뽑아 분석한 것이므로, 계절에 따른 사용률 변화 등의 영향이 끼쳤을 가능성이 있기 때문이다. 아마 월별 사용량 데이터나, 더 넓은 범위의 년도까지 분석한다면 견인 건수가 증가한 원인을 정확하게 추정할 수 있을 것이다.  

### 2-2) 가설 분석
- 견인 구역 중 횡단보도 옆에서의 견인이 제일 많을 것이다.

```python
#import pandas as pd
#import matplotlib.pylab as plt

"""from matplotlib import font_manager, rc

font_name=font_manager.FontProperties(fname="C:\Windows\Fonts\gulim.ttc").get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus']=False"""

df=pd.read_excel('./서울시 전동킥보드 견인 현황.xlsx')

df1=df.loc[0:60604,['유형']]
df1=df1.astype(str)    #string으로
df1
```

```python
import unicodedata

def count_braille_block():              #'점자블록 엘레베이터 입구' 개수 세기
    sum=0
    a='점'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c1=count_braille_block()

def count_subway():                     #'지하철역 진출입로 및 주변' 개수 세기
    sum=0
    a='지'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c2=count_subway()

def count_crosswalk():                 #'횡단보도, 산책로 등' 개수 세기
    sum=0
    a='횡'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c3=count_crosswalk()

def count_road():                      #'보도와 차도가 구분된 도로의 차도' 개수 세기
    sum=0
    a='보도와'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:3]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c4=count_road()

def count_sidewalk():                     #'보도 중앙' 개수 세기
    sum=0
    a='보도 '
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:3]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c5=count_sidewalk()

def count_bus_stop():                    #'버스정류장, 택시 승강장' 개수 세기
    sum=0
    a='버'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c6=count_bus_stop()

def count_building():                     #'건물 상가 보행자 진출입' 개수 세기
    sum=0
    a='건'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c7=count_building()

def count_bike():                        #'자전거 도로' 개수 세기
    sum=0
    a='자'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c8=count_bike()

def count_mistake():                     #'오신고' 개수 세기
    sum=0
    a='오'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c9=count_mistake()

def count_otherwise():                    #'기타' 개수 세기
    sum=0
    a='기'
    uni1=unicodedata.normalize('NFC',a)
    for i in range(0,60605):
        x=df1['유형'].loc[i]
        y=x[0:1]
        if y==uni1:
            sum+=1
        else:
            continue
    return sum

c10=count_otherwise()
```

이는 견인 구역별 견인 건수를 세는 함수를 작성한 것이다.  

견인 구역별 견인 건수를 세기 위해, '견인 유형'열의 앞 글자만 따서 개수를 세려고 하였다. 하지만 눈으로 볼 때는 엑셀 표 안에 들어 있는 한글과 주피터 노트북에서 코드를 쓸 때 적은 한글이 같지만, 직접 실행해본 결과 다르다고 인식하였다. 'UTF-8'으로 출력해본 결과, 우리 눈에 보이는 두 값이 표면은 같지만, 서로 다른 내용을 저장하고 있다는 것을 확인할 수 있었다.  

둘을 같게 만들어주기 위해 하나하나의 자음과 모음을 한글 문자열로 결합하는 'NFC' 방법을 활용해보았다. NFC를 사용하기 위해 'import unicodedata'를 선언하고 unicodedata.normalize를 사용하였다. 이를 이용하여 각각 견인 구역별 견인 수를 세는 함수를 만들었다.  

앞 글자가 겹치지 않은 유형들은 slicing을 통해 1까지만 분석하였고, 앞 글자가 겹친 유형들은 차이가 나는 지점까지 slicing을 하여 비교해주었다. 

```python
m=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10]
n=['점자블록','지하철역','횡단보도.산책로','차도','보도 중앙','버스.택시승강장',
   '건물 출입구','자전거도로','오신고','기타']

fig, ax = plt.subplots(figsize=(7,5))

plt.bar(n,m,color='indianred',width=0.6)
plt.xticks(rotation=90)
plt.xlabel('<견인구역별>',size=12,loc='left')
plt.ylabel('<견인건수>',size=12)

plt.title('<견인 구역별 견인건수>',size=15)
plt.grid(True,axis='y')                      #y축 gird만
```

이는 가설 2-1을 증명할 때와 같은 엑셀 파일에 들어있는 데이터를 사용하였으므로, 이 또한 2021년 6월부터 2022년 7월까지의 데이터를 분석한 것이다. x축에는 견인 구역별을, y축에는 견인 건수를 적용하여 그래프를 표현하였다.   

'보도와 차도가 구분된 도로의 차도'에서의 견인 건수가 제일 높았으며, 횡단보도가 2위, 버스나 택시승강장 3위로 나왔다. 1위인 보도와 차도가 구분된 도로의 차도에 견인 건수가 압도적으로 높은 것을 보아, 아마 사람들이 전동킥보드를 차도에서 이용하며 타고 오다가, 목적지에 거의 도달해서 보도까지 올리는 시간을 아껴 차도에 주차하고 가는 것이라고 추정이 된다. 물론 차도에 타는 것을 권장하고 있기는 하지만, 차도에서 타는 것도 솔직히 매우 위험한 일이다. 즉 자전거 도로처럼 전동킥보드 도로가 필요하며, 같이 혼용될수도 있기에, 개인형 이동장치 도로가 많이 생겨야 할 것이다.  

또한 차도 말고 다른 구역에서도 추가 사고가 발생할 수 있다. "점자블록"같은 곳은 특히나 더 조심히 해야한다. 따라서 탑승자들은 이에 대해 알고, 자신들의 편의를 위해 무심코 주차했던 곳에서의 추가 사고 발생까지 생각하여 올바른 곳에 주차해야한다. 

따라서 "견인 구역 중 횡단보도 옆에서의 견인이 가장 많을 것이다."라는 가설은 틀렸으며, "보도와 차도가 구분된 도로의 차도"에서의 견인이 가장 많다는 것을 알아내었다. 이 또한, 견인 구역별 추가 사고 건수 데이터까지 찾아 분석한다면, 더욱 정확한 주차 구역 지정에 대한 필요성을 더 크게 느낄 수 있을 것이다.
